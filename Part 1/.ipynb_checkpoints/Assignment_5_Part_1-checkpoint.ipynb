# Emphasising correlation

## Complete the code below to plot a scatterplot for MPG vs Weight.

Furthermore, indicate using colours the different origins of each point.

```python
fig, ax = plt.subplots(figsize=(7,6))

colors = {'USA':'tab:blue', 'JAPAN':'tab:red', 'EUROPE':'tab:green'} 

US = df[df.origin == "USA"] 
JP = df[df.origin == "JAPAN"] 
EU = df[df.origin == "EUROPE"] 


ax.scatter(EU["weight"],EU["mpg"], color = colors['EUROPE'], label = "Europe")
ax.scatter(JP["weight"],JP["mpg"], color = colors['JAPAN'], label = "Japan")
ax.scatter(US["weight"],US["mpg"], color = colors['USA'], label = "USA")

ax.legend()

ax.set_xlabel("Weight (lb)")
ax.set_ylabel("Miles per Gallon (mpg)")
ax.set_title("Scatterplot of MPG vs Weight")
plt.show()
```

### Does MPG and Weight have positive, negative or no correlation

Negative Correlation


## Complete the code below to plot a scatterplot of MPG vs Acceleration

```python
fig, ax = plt.subplots(figsize=(7,6))

colors = {'USA':'tab:blue', 'JAPAN':'tab:red', 'EUROPE':'tab:green'}


US = df[df.origin == "USA"] 
JP = df[df.origin == "JAPAN"] 
EU = df[df.origin == "EUROPE"] 


ax.scatter(EU["acceleration"],EU["mpg"], color = colors['EUROPE'], label = "Europe")
ax.scatter(JP["acceleration"],JP["mpg"], color = colors['JAPAN'], label = "Japan")
ax.scatter(US["acceleration"],US["mpg"], color = colors['USA'], label = "USA")

ax.legend()

ax.set_xlabel("Acceleration") 
ax.set_ylabel("Miles per Gallon (mpg)")
ax.set_title("Scatterplot of MPG vs Acceleration")
plt.show()


plt.show()
```

# Visualising categoricals features

## Plot a pie chart of model years
Clearly indicate each segements' numerical percentage.

```python
fig, ax = plt.subplots(figsize=(7,6))

group = df.groupby('model_year').size()

plt.pie(group, labels = group.index, shadow = True, autopct = '%1.1f%%')

plt.title("Pie Chart of Model Years")

plt.show()
```

## Plot a pie chart of origin
Clearly indicate each segements' numerical percentage.

```python
fig, ax = plt.subplots(figsize=(7,6))

group_or = df.groupby('origin').size()

plt.pie(group_or, labels = group_or.index, shadow = True, autopct = '%1.1f%%')

plt.title("Pie Chart of Vehicle Origin")

plt.show()
```

## Why is using pie charts bad when analysing categorical features?
Hint: This drawback is especially predominant in the first pie chart.

Pie charts are ineffective at displaying categorical data when there are a large number of categories. More slices in the pie, reduce readability at a glance. 



## It is better to rather use barplots when visualising categorical features. 
Therefore once again plot the model year and origin on two seperate graphs.
But instead using barplots this time.

```python
fig, ax = plt.subplots(figsize=(7,6))

ax.bar(df['model_year'],df.index)

plt.title('Bar Chart of Vehicles Year Model')
plt.xlabel('Year (19..)')
plt.ylabel('Amount of vehicles')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7,6))

ax.bar(df['origin'],df.index)

plt.title('Bar Chart of Vehicles Origin')
plt.xlabel('Origin')
plt.ylabel('Amount of vehicles')


plt.show()
```

```python
# Load in data
df = pd.read_csv('src/temps.csv', index_col=0, parse_dates=True, header=0, squeeze=True)
df = pd.DataFrame(df)

df
```

## Plot the daily temperature for each recorded measurement.

```python
fig, ax = plt.subplots(figsize=(12,5))

ax.plot(df.index,df['Temp'])

ax.set_xlabel('Date') 

ax.set_ylabel('Temperature')

ax.set_title('Daily Temperatures from 1981 to 1991')  

plt.show()
```

## Comparing different years
Complete the code below to plot the temperature of each year on a different subplot.

## Comparing distributions of time intervals
To properly visualise and compare the distributions of yearly or monthly data, we employ the use of boxplots.

### Write code below to plot a boxplot for the temperatures of each respective year.

```python
fig, ax = plt.subplots(figsize=(10,5))

ax.boxplot([year_r1['Temp'],year_r2['Temp'],year_r3['Temp'],year_r4['Temp'],year_r5['Temp'],year_r6['Temp'],year_r7['Temp'],year_r8['Temp'],year_r9['Temp']])

ax.set_xticklabels(["1981 - 1982","1982 - 1983","1983 - 1984","1984 - 1985","1985 - 1986","1986 - 1987","1987 - 1988","1988 - 1989","1989 - 1990"], rotation = 90)

ax.set_ylabel("Temperature")

ax.set_title("Boxplot of Temperatures every year from 1981 to 1991")
plt.tight_layout()
plt.show()
```
